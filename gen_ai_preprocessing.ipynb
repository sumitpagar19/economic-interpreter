```python
import pandas as pd
```

```python
df = pd.read_csv("/content/WDICSV.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
key_indicators = [
    "NY.GDP.MKTP.KD.ZG",
    "FP.CPI.TOTL.ZG",
    "SL.UEM.TOTL.ZS"
]

df = df[df["Indicator Code"].isin(key_indicators)]
```

```python
df_long = df.melt(
    id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
    var_name="Year",
    value_name="Value"
)
```

```python
df_long = df_long[df_long["Year"].str.isnumeric()]
df_long["Year"] = df_long["Year"].astype(int)
```

```python
aggregate_codes = ["WLD", "HIC", "LIC", "MIC", "LMY", "UMC"]
df_long = df_long[~df_long["Country Code"].isin(aggregate_codes)]
```

```python
df_long = df_long.sort_values(
    ["Country Name", "Indicator Code", "Year"]
)

df_long["YoY_Change"] = df_long.groupby(
    ["Country Name", "Indicator Code"]
)["Value"].diff()
```

```python
df_long["Significant_Shift"] = df_long["YoY_Change"].abs() > 5
```

```python
df_long.to_csv("wdi_cleaned.csv", index=False)
```
